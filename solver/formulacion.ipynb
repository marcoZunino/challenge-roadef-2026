## Decisions



### Routing: <br>
$x_{ij}^{dt} \in \mathbb{B}$, indicates if the segment path of $d \in D$ contains segment $(i,j) \in V \times V$ at time step $t \in T$


---

## Constraints

### Flow-conservation constraints: <br>
$$
    \sum_{j \in V \setminus \{i\}} x_{ij}^{d\tau} - \sum_{j \in V \setminus \{i\}} x_{ji}^{d\tau} =
    \begin{cases}
        1,  \ & i = s\\
        -1, \ & i = t\\
        0,  \ & i \notin \{s,t\}
    \end{cases}
    \ \\
    \ \\
    \forall i \in V, \ \forall d = (s,t) \in D, \ \forall \tau \in T
$$


#### Alternative formulation to avoid cycles: <br>

$$
    \begin{cases}
        \ \sum_{j \in V \setminus \{s\}} x_{sj}^{d\tau} = 1, \quad
        \ \sum_{j \in V \setminus \{s\}} x_{js}^{d\tau} = 0, \quad \\
        \ \sum_{j \in V \setminus \{t\}} x_{jt}^{d\tau} = 1, \quad
        \ \sum_{j \in V \setminus \{t\}} x_{tj}^{d\tau} = 0, \quad
    \end{cases}
    \forall d = (s,t) \in D, \ \forall \tau \in T
$$

$$
    \sum_{j \in V \setminus \{i\}} x_{ij}^{d\tau} = 
    \sum_{j \in V \setminus \{i\}} x_{ji}^{d\tau}, \qquad
    \forall i \in V \setminus\{s,t\}, \ \forall d = (s,t) \in D, \ \forall \tau \in T
$$

$$
    \sum_{j \in V \setminus \{i\}} x_{ij}^{d\tau} \leq 1, \qquad
    \sum_{j \in V \setminus \{i\}} x_{ji}^{d\tau} \leq 1, \qquad
    \forall i \in V \setminus\{s,t\}, \ \forall d = (s,t) \in D, \ \forall \tau \in T
$$

---

### Number of segments: <br>
$$
    \sum_{(i,j) \in V \times V} x_{ij}^{d\tau} \leq \texttt{maxSeg}, \quad
    \forall d \in D, \ \forall \tau \in T
$$


---

### Load constraints: <br>
$$
    \sum_{d \in D} \sum_{(i,j) \in V \times V} r(i,j,a,\tau) \nu (d,\tau) x_{ij}^{d\tau} \leq \lambda(a,\tau) c(a), \quad
    \forall a \in A \setminus q(\tau), \ \forall \tau \in T
$$


Caso de 1 segmento por demanda, es decir, $x_{st}^{d\tau}=1 \ \forall d=(s,t)\in D, \ \forall \tau\in T$:
$$
    \sum_{d=(s,t) \in D} r(s,t,a,\tau) \nu (d,\tau) \leq \lambda(a,\tau) c(a), \quad
    \forall a \in A \setminus q(\tau), \ \forall \tau \in T
$$

---

### Budget constraints: <br>
$$
    \sum_{d \in D} \sum_{(i,j) \in V \times V, i\neq j} |x_{ij}^{d\tau} - x_{ij}^{d,\tau-1}| \leq \kappa(\tau), \quad
    \forall \tau \in T^*
$$

#### Quadratic alternative: <br>
$$
    \sum_{d \in D} \sum_{(i,j) \in V \times V, i\neq j} \left(x_{ij}^{d\tau} - x_{ij}^{d,\tau-1}\right)^2 \leq \kappa(\tau), \quad
    \forall \tau \in T^*
$$
